# Sales Analysis Dashboard

This notebook contains exploratory data analysis of sales data.

Focuses on:
- regional performance
- category profitability
- sales trends over time
- customer segmentation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/cleaned/clean_sales.csv)
```

## 1. Sales by Region

Analyzes which regions generate the highest revenue.

```python
sales_by_region = df.groupby("region")["sales"].sum().sort_values()

sales_by_region.plot(kind="bar")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.show()
```

## 2. Profit by Category

This section shows which product categories are truly profitable.

```python
profit_by_category = df.groupby("category")["profit"].sum()

profit_by_category.plot(kind="bar")
plt.title("Total Profit by Category")
plt.ylabel("Profit")
plt.show()
```

## 3. Sales Over Time

Checks whether sales show trends, seasonality or spikes.

```python
df["order_date"] = pd.to_datetime(df["order_date"])

sales_time = df.groupby("order_date")["sales"].sum()

sales_time.plot(figsize=(10,5))
plt.title("Sales Over Time")
plt.ylabel("Sales")
plt.show()
```

## 4. Profit vs Loss Orders

Analyzes how many orders generate profit vs loss.

```python
losses = df[df["profit"] < 0]["profit"].count()
profits = df[df["profit"] >= 0]["profit"].count()

plt.pie(
    [profits, losses],
    labels=["Profit Orders", "Loss Orders"],
    autopct="%1.1f%%"
)
plt.title("Profit vs Loss Orders")
plt.show()
```

## 5. Top Customers

Identifies the most valuable customers by total sales.

```python
top_customers = df.groupby("customer_name")["sales"].sum().nlargest(10)

top_customers.plot(kind="barh")
plt.title("Top 10 Customers by Sales")
plt.show()
```
